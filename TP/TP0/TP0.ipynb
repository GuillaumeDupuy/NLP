## Exercice 3 Disponibilité des vélibs

Q1 Importer les données sous la forme d’un dataFrame
velib_t1.txt  - avec les données des stations à un instant t
velib_t2.txt  - avec les données des stations à un instant t+1

```python
import json

with open(cwd + '/data/velib_t1.txt', 'r') as file:
    data = json.load(file)

velib_t1 = pd.json_normalize(data)

with open(cwd + '/data/velib_t2.txt', 'r') as file:
    data = json.load(file)

velib_t2 = pd.json_normalize(data)
```

```python
velib_t1.head(2)
```

```python
velib_t2.head(2)
```

Q2 Représenter la localisation des stations vélib dans Paris représenter les stations en fonction du nombre de places avec un gradient

```python
import folium
from folium import plugins
import branca.colormap as cm

m = folium.Map(location=[48.864716, 2.349014], zoom_start=13)

colormap = cm.LinearColormap(colors=['green', 'yellow', 'red'], vmin=velib_t1['bike_stands'].min(), vmax=velib_t1['bike_stands'].max())


for index, station in velib_t1.iterrows():
    lat, lon = station["position.lat"], station["position.lng"]
    bike_stands = station["bike_stands"]

    color = colormap(bike_stands)

    folium.Marker(location=[lat, lon], tooltip=f"{station['name']} - {bike_stands} places",
                  icon=folium.Icon(color='white', icon_color=color)).add_to(m)

colormap.caption = 'Number of Bike Stands'
m.add_child(colormap)

m
```

Q3 Comparer pour une station donnée l’évolution de la disponibilité (en fusionnant les deux bases t et t+1)
- représenter les stations qui ont connu une évolution significative (plus de 5 changements) avec un gradient de couleurs

```python
merged_df = pd.merge(velib_t1, velib_t2, on='number', suffixes=('_t', '_t+1'))

merged_df['availability_diff'] = merged_df['bike_stands_t+1'] - merged_df['bike_stands_t']

significant_changes = merged_df[abs(merged_df['availability_diff']) > 5]

m = folium.Map(location=[48.864716, 2.349014], zoom_start=13)

colormap = cm.LinearColormap(colors=['blue', 'white', 'red'], vmin=-10, vmax=10)

for index, station in significant_changes.iterrows():
    lat, lon = station["position.lat_t"], station["position.lng_t"]
    availability_diff = station["availability_diff"]

    color = colormap(availability_diff)

    folium.Marker(
        location=[lat, lon],
        tooltip=f"{station['name_t']} - Change: {availability_diff} places",
        icon=folium.Icon(color='white', icon_color=color)
    ).add_to(m)

colormap.caption = 'Change in Bike Stand Availability'
m.add_child(colormap)

m
```
